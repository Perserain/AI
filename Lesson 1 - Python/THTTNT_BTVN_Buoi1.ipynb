```python
#1. Viết hàm trộn 2 mảng một chiều thành 1 mảng một chiều với mỗi phần tử của mảng mới là min của 2 phần tương ứng từ 2 mảng cho trước
#Trong quá trình trộng 2 mảng nếu mảng nào còn phần tử thì các phần tử còn lại của mảng đó sẽ đưa vào mảng mới
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
min = len(b) if len(a) > len(b) else len (a)
c = a if len(a) > len(b) else b
for i in range(min):
  c[i] = a[i] + b[i]
print(c)
```

```python
#2.
import random
dong = int(input("Nhập m dòng: "))
cot = int(input("Nhập n cột: "))
a = [[random.randint(-100,100) for i in range (cot)] for i in range(dong)]
tong = 0
for e in a:
  print(e)
#a) TÍNH TỔNG CÁC PHẦN TỬ THUỘC TAM GIÁC TRÊN CỦA ĐƯỜNG CHÉO PHỤ KỂ CẢ ĐƯỜNG CHÉO PHỤ TRONG MA TRẬN VUÔNG a CẤP n
if len(a) == cot:
  for i in range(len(a)):
    for j in range(cot):
      if i+j<=len(a)-1:
        tong+= a[i][j]
  print("Tổng các phần tử thuộc tam giác trên của đường chéo phụ và đường chéo phụ trong ma trận vuông a cấp n là:", tong)
else:
  print("Không phải ma trận vuông")

#b) CHUYỂN CÁC PHẦN TỬ ÂM THÀNH TRỊ TUYỆT ĐỐI CỦA NÓ TRONG a
for i in range(len(a)):
  for j in range(cot):
    if a[i][j]<0:
      a[i][j] = abs(a[i][j])
print("Ma trận sau khi chuyển các phần tử âm thành trị tuyệt đối trong a: ")
for e in a:
  print(e)

#c) THAY CÁC PHẦN TỬ CHẴN TRONG a BẰNG SỐ NGUYÊN x CHO TRƯỚC
x = int(input("Nhập x để thay thế các phần tử chẵn trong a: "))
for i in range(len(a)):
  for j in range(cot):
    if a[i][j] %2==0:
      a[i][j] = x
print("Ma trận sau khi thay các phần tử chẵn trong a bằng số nguyên", x, "cho trước: ")
for e in a:
  print(e)
#d) KIỂM TRA a CÓ TOÀN CHẴN KHÔNG?
check = 1
for i in range(len(a)):
  for j in range(cot):
    if a[i][j] %2!=0:
      check -=1
if check ==1:
  print("Ma trận a toàn chẵn")
else:
  print("Ma trận a không toàn chẵn")

#e) CHO MA TRẬN VUÔNG a CẤP n, VIẾT HÀM KIỂM TRA a CÓ ĐỐI XỨNG KHÔNG. BIẾT RẰNG MA TRẬN ĐỐI XỨNG LÀ MA TRẬN CÓ a[i][j] = a[j][i]
if len(a) == cot:
  check = 1
  for i in range(len(a)):
    for j in range(cot):
      if a[i][j] != a[j][i]:
        check = 0;
        break
  if check==1:
    print("Ma trận đối xứng")
  else:
    print("Ma trận không đối xứng")
else:
  print("Không phải ma trận vuông")
#f) KIỂM TRA MA TRẬN VUÔNG a CẤP n CÓ ĐƯỜNG CHÉO CHÍNH TĂNG DẦN KHÔNG?
if len(a) == cot:
  check = 1
  for i in range(len(a)-1):
    if a[i][i] > a[i+1][i+1]:
      check = 0
      break
  if check == 1:
    print("Ma trận có đường chéo chính tăng dần")
  else:
    print("Ma trận không có đường chéo chính tăng dần")
else:
  print("Không phải ma trận vuông")
#g) XUẤT CÁC PHẦN TỬ THUỘC TAM GIÁC DƯỚI CỦA ĐƯỜNG CHÉO PHỤ KỂ CẢ ĐƯỜNG CHÉO PHỤ TRONG MA TRẬN VUÔNG a CẤP n
if len(a) == cot:
  print("Các phần tử thuộc tam giác dưới của đường chéo phụ kể cả đường chéo phụ là: ")
  for i in range(len(a)):
    for j in range(cot):
      if i+j>=len(a)-1:
        print("a[", i, "][",j,"] =",a[i][j])
else:
  print("Không phải ma trận vuông")
#h) KIỂM TRA MA TRẬN VUÔNG a CẤP n CÓ ĐƯỜNG CHÉO PHỤ GIẢM DẦN KHÔNG?
if len(a) == cot:
  check = 1
  duong_cheo_phu = []
  for i in range(len(a)):
    duong_cheo_phu.append(a[i][len(a)-1-i])

  for i in range(len(duong_cheo_phu) - 1):
    if duong_cheo_phu[i] < duong_cheo_phu[i+1]:
      check = 0
      break

  if check == 1:
    print("Ma trận có đường chéo phụ giảm dần")
  else:
    print("Ma trận không có đường chéo phụ giảm dần")
else:
  print("Không phải ma trận vuông")
```
