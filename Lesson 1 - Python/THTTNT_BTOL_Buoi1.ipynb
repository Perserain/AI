```python
#14. Cho mảng 2 chiều a có m dòng, n cột chứa số nguyên
#a) Tạo mảng a chứa các số nguyên ngẫu nhiên
import numpy as np
import random
m = int(input("Nhập dòng: ")) #Dòng
n = int(input("Nhập cột: ")) #Cột
a = [[random.randint(1,100) for i in range (n)] for i in range(m)]
a = np.array(a)
print(a)

#b) Xuất các phần tử thuộc dòng k
k = int(input("Dòng muốn xuất: ")) #Nhập dòng k vào
print(a[k])

#c) Xuất các phần tử thuộc cột k
k = int(input("Cột muốn xuất: "))
for i in range(len(a)):
  print(a[i][k])

#d) Tìm dòng có tổng lớn nhất (45??)
dongTongLonNhat = 0
max = np.sum(a[0])
for i in range(1, len(a)):
  tong = np.sum(a[i])
  if(tong>max):
    dongLonNhat = i
    max = tong
print(f"Dòng lớn nhất là: {dongTongLonNhat}")

#e) Tìm cột có tích nhỏ nhất
cotTichNhoNhat = -1
min = -1
for i in range(len(a[0])):
  tich = 1
  for j in range(len(a)):
    tich *= a[j][i]
  if(tich<min or min==-1):
    cotTichNhoNhat = i
print(f"Cột có tích nhỏ nhất là: {cotTichNhoNhat}")

#f) Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a
print("Các phần tử thuộc dòng chẵn và cột lẻ trong a là: ")
for i in range(len(a)):
  for j in range(len(a[i])):
    if(i%2==0 and j%2!=0):
      print(a[i][j])

#g) Tính trung bình cộng của các phần tử chẵn thuộc dòng lẻ của a
tong = 0
soluong = 0
for i in range(len(a)):
  if(i%2!=0):
    for j in range(len(a[i])):
      if(a[i][j]%2==0):
        tong += a[i][j]
        soluong += 1
tbc = tong/soluong
print(f"Trung bình cộng của các phần tử chẵn thuộc dòng lẻ của a là: {tbc}")

#h) Tính trung bình cộng các phần tử thuộc biên
tong = 0
soluong = 0
for i in range(len(a)):
  for j in range(len(a[i])):
    if(i==0 or j==0 or i==len(a)-1 or j==len(a[i])-1):
      tong += a[i][j]
      soluong +=1
tbc = tong/soluong
print(f"Trung bình cộng các phần tử thuộc biên là: {tbc}")

# Tính trung bình cộng các phần tử không thuộc biên
tong = 0
soluong = 0
for i in range(len(a)):
  for j in range(len(a[i])):
    if(i!=0 and j!=0 and i!=len(a)-1 and j!=len(a[i])-1):
      tong += a[i][j]
      soluong +=1
tbc = tong/soluong
print(f"Trung bình cộng các phần tử không thuộc biên là: {tbc}")
```
