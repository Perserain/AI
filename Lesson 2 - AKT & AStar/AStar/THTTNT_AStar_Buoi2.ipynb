#BTOL: Thuật toán A* tìm 2 đường bất kỳ trong đồ thị

```python
from collections import deque

class Graph:
  def __init__(self, adjacency_list):
    self.adjacency_list = adjacency_list

  def get_neighbors(self, v): #Lấy các node cận kề node hiện tại
    return self.adjacency_list[v]

  def h(self, n): #Hàm tính heuristic
    H = {
      'A': 1,
      'B': 1,
      'C': 1,
      'D': 1
    }

    return H[n]

  def a_star_algorithm(self, start_node, stop_node): #Thuật toán A*, truyền vào 2 tham số node bắt đầu và node đích
    open_list = set([start_node]) #Bước 1: Xét node bắt dầu
    closed_list = set([]) #Bước 1: closed_list rỗng

    poo = {}
    poo[start_node] = 0 #g[n]

    par = {}
    par[start_node] = start_node #lưu dấu vết

    while len(open_list) > 0: #Chạy cho đến khi không còn node nào để xét nữa
      n = None

      for v in open_list:
        if n == None or poo[v] + self.h(v) < poo[n] + self.h(n): #tìm node có chi phí f[n] min
          n = v         #sau khi tìm gán n = node có chi phí f[n] min

      if n == None:
        print('Path does not exist!')
        return None

      if n == stop_node: #sau khi tìm thấy đích
        reconst_path = []

        while par[n] != n:
          reconst_path.append(n)
          n = par[n] #truy vết

        reconst_path.append(start_node)

        reconst_path.reverse() #đảo ngược lại từ đích tới đầu

        print('Path found: {}'.format(reconst_path))
        return reconst_path #trả lại kết quả

      for(m, weight) in self.get_neighbors(n): #trọng số các node cận kề
        if m not in open_list and m not in closed_list: #nếu m chưa được xét
          open_list.add(m) #thêm m vào open_list để xét
          par[m] = n #truy vết
          poo[m] = poo[n] + weight #g[m] = g[n] + trọng số
        else:
          if poo[m] > poo[n] + weight:
            poo[m] = poo[n] + weight
            par[m] = n #lấy chi phí min

            if m in closed_list:
              closed_list.remove(m)
              open_list.add(m)
      open_list.remove(n)
      closed_list.add(n)
    print('Path does not exist!')
    return None

adjacancy_list = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)]
}

graph1= Graph(adjacancy_list)
graph1.a_star_algorithm('A', 'D') #Start_node: A, Stop_node: D
```

#CÀI ĐẶT BÀI TOÁN NGƯỜI GIAO HÀNG THEO A*

```python
import heapq

def a_star_tsp(matrix, start_node=0):
    n = len(matrix)
    # Hàng đợi ưu tiên: (f, g, current_node, visited_mask)
    # visited_mask là bitmask để quản lý các thành phố đã đi qua
    initial_visited = 1 << start_node
    queue = [(0, 0, start_node, [start_node])]

    def heuristic(current, visited_indices):
        # Hàm Heuristic đơn giản: khoảng cách nhỏ nhất từ node hiện tại đến một node chưa thăm
        unvisited = [i for i in range(n) if i not in visited_indices]
        if not unvisited:
            return matrix[current][start_node] # Quay về điểm đầu
        return min(matrix[current][i] for i in unvisited)

    best_costs = {}

    while queue:
        f, g, u, path = heapq.heappop(queue)

        if len(path) == n:
            final_cost = g + matrix[u][start_node]
            full_path = path + [start_node]
            print(f"Path found: {full_path} with total cost: {final_cost}")
            return full_path, final_cost

        state = (u, tuple(sorted(path)))
        if state in best_costs and best_costs[state] <= g:
            continue
        best_costs[state] = g

        for v in range(n):
            if v not in path:
                new_g = g + matrix[u][v]
                new_path = path + [v]
                h = heuristic(v, new_path)
                heapq.heappush(queue, (new_g + h, new_g, v, new_path))

    return None

# Ma trận khoảng cách giữa 4 thành phố A(0), B(1), C(2), D(3)
# Ví dụ: matrix[0][1] là khoảng cách từ A đến B
distance_matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

path, cost = a_star_tsp(distance_matrix, 0)
```
